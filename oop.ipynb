```python
# Python programming
class Man:
 """functions of class """
 def __init__(self, name):
        self.name = name
        print("Initilized!")
 def hello(self):
        print("Hello " + self.name + "!")
 def goodbye(self):
        print("Good-bye " + self.name + "!")
```

```python
m = Man("David")
```

```python
m.hello()
```

```python
m.goodbye()
```

```python
m.goodbye()
```
